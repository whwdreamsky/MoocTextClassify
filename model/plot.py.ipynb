```python
import sklearn
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
```

```python
ffeature = open('feature_uniq.input','r')
corpus = []
for line in ffeature:
    tmp = line.strip().split('\t')
    if len(tmp)<3:
        continue
    corpus.append(tmp)
corpus = np.array(corpus)
X_train = corpus[:,1:]
X_train = X_train.astype('float')
Y_train = corpus[:,0]
```

```python
f2  = plt.figure('LSI')
labels = ['contradictory','correct','partially_correct_incomplete','non_domain','irrelevant']
cs = ['r','b','g','m','y']
for i in range(len(labels)):
    x = X_train[np.where(Y_train==labels[i])]
    plt.scatter(x[:,0],x[:,1],color=cs[i])
plt.show()
```
